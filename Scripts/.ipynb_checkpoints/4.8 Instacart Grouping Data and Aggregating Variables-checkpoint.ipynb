**The Steps followed in answering this question is:**

```python
# Viewing the merged dataset which contain the loyalty flag
ords_prods_merge[['user_id','loyalty_flag','order_number','prices']].head(100)
```

```python
# applying the aggregation functions logic for the whole data set
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

**Based on the above statistics, we can observe 2 points:**

**1)The new customers have a higher purchasing power than the Loyal and Regular customers.**

**2)The min and max values are set by the data engineer to be 1.0 and 99999.0 respectively. We need to do further analysis to check which product/s are skewing the dataset prices column.**

```python
# Checking the mean value
ords_prods_merge.groupby('loyalty_flag')['prices'].mean()
```

```python
#Lets investigate with the descriptive statistics
ords_prods_merge.info()
```

```python
#Running the describe function on the prices column of whole dataset
ords_prods_merge['prices'].describe()
```

**While viewing the mean value, we can see that the mean value of the prices column is 11.9 with the maximum value being 99999.0. This might be an anomaly. Lets check which products have those maximum values.**

```python
#First, lets check the records with highest prices
ords_prods_merge['prices'].sort_values(ascending = False)
```

```python
# Checking the Unique column values
ords_prods_merge['prices'].unique()
```

**The two values which are of anamoly are 99999.0 and 14900.0. These are higher than all the other prices, which can be seen from the above Unique values in the Prices.**

```python
#Products of the two highest priced products
two_max = ords_prods_merge.loc[ords_prods_merge['prices'].isin([99999.0,14900.0])]
```

```python
two_max.head()
```

```python
two_max['product_name'].unique()
```

```python
# Viewing the 2 products individually 
two_max[two_max['product_name'] == '2 % Reduced Fat  Milk']
```

```python
two_max[two_max['product_name'] == 'Lowfat 2% Milkfat Cottage Cheese']
```

```python
#Lets filter the main data set by taking out the 2 weird priced products
ords_prods_merge_dept = ords_prods_merge.loc[ords_prods_merge['department_id'].isin([16])]
ords_prods_merge_new = ords_prods_merge_dept.loc[~ords_prods_merge_dept['prices'].isin([99999.0,14900.0])]
```

```python
ords_prods_merge_new.head()
```

```python
ords_prods_merge_new['prices'].describe()
```

**The mean value of the dataset without the Inconsistent Cottage Cheese prices are 8.33. I will impute the values by using replace function.**

```python
#Lets impute the value of both the products as the inconsistencies for the prices are wrong.
ords_prods_merge['prices'] = ords_prods_merge['prices'].replace({99999: 10, 14900: 8.33})
```

```python
ords_prods_merge['prices'].describe()
```

```python
# Now, performing the Spending habits for the above dataset
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

**We can see that the New Customer Spending habits are bit more than the Loyal and Regular customers.**

## Question 6

**Create a spending flag for each user based on the average price across all their orders.**

```python
# Creating an average price column by performing transform function
ords_prods_merge['average_price'] = ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
```

```python
#Checking the new column
ords_prods_merge.head()
```

```python
# Creating the Spending flag logic
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10 , 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10 , 'spending_flag'] = 'High Spender'
```

```python
#Checking the Spending flag 
ords_prods_merge.head(100)
```

```python
#Checking the value counts of the Spending flag
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## Question 7

**Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column.**

```python
#Creating the new column by transforming the median days since prior order
ords_prods_merge['prior_order_median'] = ords_prods_merge.groupby('user_id')['days_since_last_order'].transform(np.median)
```

```python
#Checking the new column
ords_prods_merge.head()
```

```python
#Creating the order frequency flag
ords_prods_merge.loc[ords_prods_merge['prior_order_median'] > 20 , 'order_frequency_flag'] = 'Non-frequent customer.'
```

```python
ords_prods_merge.loc[(ords_prods_merge['prior_order_median'] > 10) & (ords_prods_merge['prior_order_median'] <= 20 ) , 'order_frequency_flag'] = 'Regular customer.'
```

```python
ords_prods_merge.loc[ords_prods_merge['prior_order_median'] <= 10 , 'order_frequency_flag'] = 'Frequent customer.'
```

```python
# Checking the new flag in the dataset
ords_prods_merge.head()
```

```python
# Checking the value counts of the order_frequency_flag 
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
#Viewing the null values
ords_prods_merge.loc[ords_prods_merge['order_frequency_flag'].isnull()]
```

**We can see that the order_frequency flag with NaN is for new customer with user_id 159838. Hence, the above 5 rows need not be imputed**
