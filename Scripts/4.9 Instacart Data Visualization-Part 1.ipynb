```python
# Mixing Type values
for col in customers.columns.tolist():
    weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (customers[weird]) > 0:
        print ('Mixed type value present in: '+col)
    else:
        print('No mixed type values in the column: '+ col)
```

```python
# Lets convert the First name field to String to address mixed type values within the field
customers['First Name'] = customers['First Name'].astype('str')
```

```python
# Re-Checking Mixing Type values
for col in customers.columns.tolist():
    weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (customers[weird]) > 0:
        print ('Mixed type value present in: '+col)
    else:
        print('No mixed type values in the column: '+ col)
```
